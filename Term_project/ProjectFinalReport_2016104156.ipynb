## 주제 : 교통사고의 주원인은 무엇일까?

### (1) 주제 선정 이유 :

우리는 살아가면서 많은 위험에 직면합니다. 그중에 가장 언제 일어나도 전혀 이상하지 않은 위험은 교통사고라 말할 수 있을 것입니다. 교통사고는 우리 주변에서 수없이 일어나고 있지만, 사람들은 그 위험성을 간과한 체 살아가고 있다고 생각합니다. 예를 들어 버스를 타고 갈 때 안전벨트를 매지 않는다던가, 귀에 이어폰을 꽂고 횡단보도를 건넌다던가, 아니면 급하다는 이유로 무단횡단을 한다던가 말입니다. 저는 이러한 교통사고가 주로 일어나는 원인이 무엇인지 궁금하게 되어 이러한 주제를 선정하게 되었습니다. 

### (2) 가설 정의 : 

아무래도 교통사고가 일어나는 원인 중 가장 큰 원인은 법규위반이라 생각합니다. 그중에서도 보행자가 신호를 안 지키고 무단횡단을 하는 것과 운전자가 속도위반이나 신호 위반등 을 해서 사고가 제일 자주 난다고 예상합니다. 그래서 저는 법규위반별사고 다발지역정보와 보행자 무단횡단 사고 다발 지역 정보를 조회하여 교통사고의 원인 중 가장 많은 비율을 차지하는 원인은 무엇인지 알아 볼 예정입니다.

### (3) 인터넷을 통한 데이터 획득 : 

저는 우선 데이터를 공공데이터 포털에서 받았기에 다음과 같은 절차로 데이터를 획득하였습니다.<br>
<br>
1) 공공데이터 포털에 들어가 회원가입을 하고 <br>
2) 오픈API 중 교통사고와 관련된 파일을 검색 <br>
3) '보행자무단횡단사고다발지역정보_Rest_조회.xml' 와 '법규위반별사고다발지역정보_Rest_조회.xml' 파일 활용신청 <br>
4) 활용신청을 하고 승인을 받은 후 인증키를 받음 <br>
